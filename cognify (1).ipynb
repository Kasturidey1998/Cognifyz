```python
# Convert categorical investment preference rankings to numerical values
investment_columns = ['Mutual_Funds', 'Equity_Market', 'Debentures', 
                     'Government_Bonds', 'Fixed_Deposits', 'PPF', 'Gold']
for col in investment_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
# Convert duration to numerical (extract the lower bound)
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
df['Duration_numeric'] = df['Duration'].map(duration_mapping)
```

```python
# Convert expected returns to numerical (take the midpoint)
def convert_expect(expect_str):
    if isinstance(expect_str, str):
        if '-' in expect_str:
            low, high = expect_str.replace('%', '').split('-')
            return (float(low) + float(high)) / 2
        elif '%' in expect_str:
            return float(expect_str.replace('%', ''))
    return np.nan

df['Expected_Return_numeric'] = df['Expect'].apply(convert_expect)
```

```python
# Select numerical columns for correlation analysis
numerical_columns = ['age', 'Duration_numeric', 'Expected_Return_numeric'] + investment_columns
correlation_df = df[numerical_columns]
```

```python
# Calculate correlation matrix
correlation_matrix = correlation_df.corr()
```

```python
# Plot correlation heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt=".2f")
plt.title('Correlation Matrix of Investment Factors')
plt.tight_layout()
plt.show()
```

```python
# Pair plots for key variables
sns.pairplot(correlation_df[['age', 'Duration_numeric', 'Expected_Return_numeric', 
                            'Mutual_Funds', 'Equity_Market', 'Fixed_Deposits']])
plt.suptitle('Pair Plots of Key Variables', y=1.02)
plt.show()
```

```python
# Scatter plot of age vs expected return with duration as hue
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='Expected_Return_numeric', 
                hue='Duration', palette='viridis', s=100)
plt.title('Age vs Expected Return by Investment Duration')
plt.xlabel('Age')
plt.ylabel('Expected Return (%)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
